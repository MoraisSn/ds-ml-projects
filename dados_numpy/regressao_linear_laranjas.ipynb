```python
import numpy as np
url = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
dado = np.loadtxt(url, delimiter=',',usecols=np.arange(1,6,1),skiprows=1)
dado
```

```python
diametro_laranja = dado[:5000,0]
diametro_toranja = dado[5000:,0]
peso_laranja = dado[:5000,1]
peso_toranja = dado[5000:,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
plt.legend(['peso_laranja','peso_toranja'])
```

image.png

```python
x = diametro_laranja
y = peso_laranja
n = dado.size
a = ((n * np.sum(x * y)) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
```

image.png

```python
b = np.mean(y) - a * np.mean(x)
```

```python
y = a * x + b
```

```python
# toranja
X = diametro_toranja
Y = peso_toranja
A = ((n * np.sum(X * Y)) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)
```

```python
B = np.mean(Y) - A * np.mean(X)
```

```python
Y = A * X + B
```

```python
np.linalg.norm(diametro_laranja-y)
```

```python
np.linalg.norm(diametro_toranja-Y)
```

```python
plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
plt.plot(x, y)
plt.plot(X, Y)
plt.legend(['peso_laranja','peso_toranja'])
```
